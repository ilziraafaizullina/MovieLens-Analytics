## 0) Разминочка: подключаем анализатор и загружаем данные

```python
%timeit -n 1 -r 1 1+1
from movielens_analysis import Ratings, Tags, Movies, Links

DATA_PATH = 'ml-latest-small'

ratings = Ratings(f"{DATA_PATH}/ratings.csv")
tags = Tags(f"{DATA_PATH}/tags.csv")
movies = Movies(f"{DATA_PATH}/movies.csv")
links = Links(f"{DATA_PATH}/links.csv")

('Готово', type(ratings).__name__, type(tags).__name__, type(movies).__name__, type(links).__name__)
```

## 1) Когда люди были активнее всего?
### Вопрос
Есть ли годы, когда оценок ставили заметно больше?

```python
%timeit -n 1 -r 1 ratings.movies.dist_by_year()
ratings.movies.dist_by_year()
```

### Ответ
Это распределение показывает «темп жизни» аудитории по годам: где пик — там максимум активности.
И по результатам видно, что самыми активными были люди в 1996 году. Вот они - истинные ценители фильмов!

## 2) Люди чаще хвалят или ругают?
### Вопрос
Как выглядит распределение оценок: тянет к высоким или к низким?

```python
%timeit -n 1 -r 1 ratings.movies.dist_by_rating()
ratings.movies.dist_by_rating()
```

### Ответ
Если рядом с высокой оценкой стоит большое число - люди чаще остаются довольными, если рядом с низкой оценкой стоит большое число - пользователи чаще остаются недовольными. Таким образом, мы видим, что чаще всего люди ставят именно положительную оценку, но тем не менее, второе число оценок по убыванию приходится на оценку 3.0, что говорит о том, что имеется значительное количество оценок "удовлетворительно".

## 3) Какие фильмы обсуждают больше всего?
### Вопрос
Что чаще всего получают оценки — наши «хиты по количеству голосов»?

```python
%timeit -n 1 -r 1 ratings.movies.top_by_num_of_ratings(10)
ratings.movies.top_by_num_of_ratings(10)
```

### Ответ
Это не обязательно «лучшие» фильмы — это чаще всего **самые видимые**. Популярность ≠ качество, но именно она двигает обсуждение. Число справа - количество оценок для конктретного фильма. Здесь мы видим "топ 10 самых обсуждаемых фильмов" за все года.

## 4) Что считается лучшим: среднее или медиана?
### Вопрос
Средняя оценка чувствительна к выбросам. Медиана — более «устойчивая». Что покажет топ по каждой метрике?

```python
%timeit -n 1 -r 1 ratings.movies.top_by_ratings(10, 'average')
ratings.movies.top_by_ratings(10, 'average')
```

```python
%timeit -n 1 -r 1 ratings.movies.top_by_ratings(10, 'median')
ratings.movies.top_by_ratings(10, 'median')
```

### Ответ
Итак, мы не видим различие между результатами! Повезло, повезло... Но что, если бы они были?
Если медиана сильно отличается от среднего, значит данные несимметричны и в них есть выбросы.
👉 Среднее “тянется” выбросами,
👉 медиана показывает типичное значение.
Для выводов почти всегда лучше медиана, если данные «грязные», человеческие или эмоциональные (рейтинги, доходы, оценки).
